### 20) Total costs by Branch

```python
cogs_column = 'cogs'
taxes_column = 'Tax 5%'

# Calculate total costs
df['Total Costs'] = df[cogs_column] + df[taxes_column]
# Analyze costs by branch
branch_costs = df.groupby('Branch')['Total Costs'].sum()

# Analyze costs by product line
product_line_costs = df.groupby('Product line')['Total Costs'].sum()
# Identify areas with high costs
high_cost_branches = branch_costs[branch_costs > branch_costs.mean()]

# Identify areas with high costs by product line
high_cost_product_lines = product_line_costs[product_line_costs > product_line_costs.mean()]

# Create a bar chart for branch costs
branch_costs.plot(kind='bar', title='Total Costs by Branch')
plt.show()
```

```python

mean_rating = df['Rating'].mean()
mse = ((df['Rating'] - mean_rating) ** 2).mean()
rms = np.sqrt(mse)

print(f"Root Mean Square (RMS): {rms}")
```

### Mean Absolute Error: 

```python

mean_rating = df['Rating'].mean()
mae = (df['Rating'] - mean_rating).abs().mean()

print(f"Mean Absolute Error (MAE): {mae}")
```
